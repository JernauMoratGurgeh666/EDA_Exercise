```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the crime dataframe
crime = pd.read_csv('data/KCSO_Offense_Reports__2020_to_Present.csv')

# Group the data by 'zip' and 'nibrs_code_name' and count the occurrences
crime_counts = crime.groupby(['zip', 'nibrs_code_name']).size().reset_index(name='count')

# Pivot the data to have 'nibrs_code_name' as columns and 'zip' as index
crime_pivot = crime_counts.pivot(index='zip', columns='nibrs_code_name', values='count').fillna(0)

# Calculate the total number of overall crimes for each zipcode
crime_pivot['Total'] = crime_pivot.sum(axis=1)

# Select the top 20 zipcodes with the highest number of crimes
top_20_zipcodes = crime_pivot.nlargest(20, 'Total')

# Convert the 'zip' column to integers
top_20_zipcodes.index = top_20_zipcodes.index.astype(int)

# Filter the zipcodes in the range [98001, 98199]
top_20_zipcodes = top_20_zipcodes[(top_20_zipcodes.index >= 98001) & (top_20_zipcodes.index <= 98199)]

# Plot the bar chart
plt.figure(figsize=(12, 6))
top_20_zipcodes.drop('Total', axis=1).plot(kind='bar', stacked=True, colormap='tab20', ax=plt.gca())
plt.xlabel('Zipcode')
plt.ylabel('Number of Crimes')
plt.title('Number of Crimes by Zipcode and Crime Type (Top 20 Zipcodes)')
plt.legend(title='Crime Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for readability
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Convert 'incident_datetime' column to datetime data type
crime['incident_datetime'] = pd.to_datetime(crime['incident_datetime'], errors='coerce')

# Step 2: Filter data for crimes from 2020 to 2023
crime_filtered = crime[(crime['incident_datetime'] >= '2020-01-01') & (crime['incident_datetime'] <= '2023-12-31')]

# Step 3: Group filtered data by date and count the number of crimes for each date
crime_count_by_date = crime_filtered.groupby(crime_filtered['incident_datetime'].dt.date).size().reset_index(name='crime_count')

# Step 4: Create a line plot
plt.figure(figsize=(12, 6))
plt.plot(crime_count_by_date['incident_datetime'], crime_count_by_date['crime_count'], marker='o', linestyle='-')
plt.title('Total Crime Count Over Time (2020-2023)')
plt.xlabel('Date')
plt.ylabel('Crime Count')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame with data for reported crimes and population over time
data = {
    'Year': [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023],
    'New York Crimes': [10000, 9500, 9200, 9000, 8500, 8000, 7800, 7600, 7400],
    'San Francisco Crimes': [5000, 4800, 4700, 4600, 4500, 4400, 4300, 4200, 4100],
    'LA Crimes': [8000, 7800, 7500, 7300, 7100, 6900, 6700, 6500, 6300],
    'Miami Crimes': [3000, 2900, 2800, 2700, 2600, 2500, 2400, 2300, 2200],
    'King County Crimes': [2000, 1950, 1900, 1850, 1800, 1750, 1700, 1650, 1600],
    'New York Population': [8000000, 8050000, 8100000, 8150000, 8200000, 8250000, 8300000, 8350000, 8400000],
    'San Francisco Population': [800000, 805000, 810000, 815000, 820000, 825000, 830000, 835000, 840000],
    'LA Population': [10000000, 10050000, 10100000, 10150000, 10200000, 10250000, 10300000, 10350000, 10400000],
    'Miami Population': [500000, 505000, 510000, 515000, 520000, 525000, 530000, 535000, 540000],
    'King County Population': [2200000, 2210000, 2220000, 2230000, 2240000, 2250000, 2260000, 2270000, 2280000]
}

df = pd.DataFrame(data)

# Calculate crimes per capita for each city
for city in ['New York', 'San Francisco', 'LA', 'Miami', 'King County']:
    df[f'{city} Crimes per Capita'] = df[f'{city} Crimes'] / df[f'{city} Population'] * 1000  # Per 1000 people

# Create a line chart for crimes per capita
plt.figure(figsize=(10, 6))

# Plot lines for crimes per capita for each city with distinctive colors
for city in ['New York', 'San Francisco', 'LA', 'Miami', 'King County']:
    plt.plot(df['Year'], df[f'{city} Crimes per Capita'], marker='o', label=city)

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Crimes per Capita (per 1000 people)')
plt.title('Reported Crimes per Capita in US Areas (2015 to 2023)')

# Add a legend to explain the colors
plt.legend(loc='upper right')

# Set x-axis range from 2015 to 2023
plt.xlim(2015, 2023)

# Display the plot
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import geopy.distance

# Load the housing dataset
df = pd.read_csv('data/king_county_house_JOIN_DATABASES.csv', sep=';')

# Define the approximate coordinates of the King County city center (Seattle)
city_center_lat = 47.6062
city_center_lon = -122.3321

# Create a function to calculate the distance between two coordinates
def calculate_distance(row):
    apartment_coords = (row['lat'], row['long'])
    city_center_coords = (city_center_lat, city_center_lon)
    return geopy.distance.geodesic(apartment_coords, city_center_coords).miles

# Calculate the distance to the city center for each apartment
df['distance_to_city_center'] = df.apply(calculate_distance, axis=1)

# Define the maximum distance from the city center (in miles)
max_distance = 0.9  # You can adjust this value as needed

# Filter apartments that are within the specified distance from the city center
LowCrimeCityCenter = df[df['distance_to_city_center'] <= max_distance]

# Display the remaining apartments
print(LowCrimeCityCenter)
```

```python
# Export the "LowCrimeCityCenter" data to a CSV file
LowCrimeCityCenter.to_csv('LowCrimeCityCenter.csv', index=False)

# Verify the number of rows in the exported CSV file
print("Number of rows in the exported CSV file:", len(LowCrimeCityCenter))
```
